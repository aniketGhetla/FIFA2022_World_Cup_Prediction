```python
from bs4 import BeautifulSoup
import pandas as pd

with open("/content/drive/MyDrive/FIFAWorldCup.html", 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')
tables = soup.find_all('table')

# Convert tables to DataFrames
dfs = []
for table in tables:
    dfs.append(pd.read_html(str(table))[0])
```

```python
dict_table ={}
for letter,i in zip(alphabet,range (12,68,7)):
  df = dfs[i]
  df.rename(columns={df.columns[1]:'Team'},inplace=True)
  # Check if 'Qualification' column exists before popping
  if 'Qualification' in df.columns:
    df.pop('Qualification')
  dfs.append(df)
  dict_table[f'Group {letter}']=df
```

```python
with open('dict_table', 'wb') as output: # wb= writebytes
    pickle.dump(dict_table, output)
```
